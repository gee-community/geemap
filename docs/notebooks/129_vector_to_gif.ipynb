**Creating animated GIF from vector data**

Inspired by Johannes Uhl's [shapefile2gif](https://github.com/johannesuhl/shapefile2gif), I created a `vector_to_gif()` function in geemap that makes it much easier to create animated GIF from vector data with only one line of code.


Uncomment the following line to install [geemap](https://geemap.org) if needed.

```python
# !pip install -U geemap
```

```python
import geemap
```

```python
data = 'https://github.com/giswqs/data/raw/main/us/boulder_buildings.zip'
```

```python
m = geemap.Map(center=[39.9898, -105.2532], zoom=14)
m.add_vector(data, layer_name='Buildings')
m
```

```python
out_gif = 'buildings.gif'
colname = 'year_built'
title = 'Building Evolution in Boulder, Colorado, USA (1950-2015)'
```

```python
geemap.vector_to_gif(
    data,
    out_gif,
    colname,
    vmin=1950,
    vmax=2015,
    step=10,
    facecolor='black',
    figsize=(10, 8),
    title=title,
    xy=('1%', '1%'),
    fontsize=20,
    progress_bar_color='blue',
    progress_bar_height=10,
    dpi=300,
    fps=10,
    mp4=False,
    verbose=True,
)
```
